### Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('netflix_titles.csv')
```

```python
df.columns
```

### 1. Make a visualization showing the total number of movies watched by children

```python
df['listed_in'].str.contains('Child')
```

```python
sum(df['listed_in'].str.contains('Child'))
```

```python
df[df['listed_in'].str.contains('Child')]
```

```python
sns.countplot(x=df['listed_in'].str.contains('Child'));
```

```python
child = sum(df['listed_in'].str.contains('Kid'))
```

### 2. Make a visualization showing the total number of standup comedies

```python
df.head()
```

```python
set(df['listed_in'])
```

```python
suc = df[df['listed_in'].str.contains('Stand-Up Comedy')]
```

```python
suc.shape
```

```python
plt.bar(['Child Shows','Stand-Up Comedy'],[child, len(suc)])
```

```python
plt.pie([child, len(suc)], 
        labels=['Child Shows','Stand-Up Comedy'],
       autopct='%2.2f%%', shadow=True,radius=1.2);
```

### 3. Make a visualization showing most watched shows.

```python
df.columns
```

```python
seasons = df[df.duration.str.contains('Season')]
```

```python
seasons.sort_values(by='duration', ascending=False)
```

```python
set(seasons.duration)
```

```python
seasons = seasons[seasons.duration.str.startswith('1')]
```

```python
len(seasons[seasons.duration.str.startswith('1 ')])
```

```python
ten = seasons[seasons.duration.str.startswith('10')]
eleven = seasons[seasons.duration.str.startswith('11')]
twelve = seasons[seasons.duration.str.startswith('12')]
thirteen = seasons[seasons.duration.str.startswith('13')]
fifteen = seasons[seasons.duration.str.startswith('15')]
sixteen = seasons[seasons.duration.str.startswith('16')]
```

```python
plt.bar([10,11,12,13,15,16],
        [len(ten),len(eleven),len(twelve),len(thirteen),
        len(fifteen),len(sixteen)])
```

```python
seasons = seasons[not seasons.duration.startswith('1 ')]
```

```python
sns.countplot(x=seasons.duration);
```

#### 4. Make a visualization showing highest rated show

```python
df.columns
```

```python
sns.countplot(y=df.rating)
```

### Make a dashboard (DASHBOARD A) containing all of these above visualizations.

```python
plt.figure(figsize=(16,9))
plt.subplot(2,2,1)
sns.countplot(x=df['listed_in'].str.contains('Child'));
plt.subplot(2,2,2)
plt.pie([child, len(suc)], 
        labels=['Child Shows','Stand-Up Comedy'],
       autopct='%2.2f%%', shadow=True,radius=1.2);
plt.subplot(2,2,3)
plt.bar([10,11,12,13,15,16],
        [len(ten),len(eleven),len(twelve),len(thirteen),
        len(fifteen),len(sixteen)])
plt.subplot(2,2,4)
sns.countplot(y=df.rating)
plt.savefig('myplot.png')
```
