```python
# Perform clustering of the iris dataset based on all variables using 
# Gaussian mixture models.
# Use PCA to visualize clusters.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
df
```

```python
x = df.drop('species', axis = 1)
```

### Build the model

```python
from sklearn.mixture import GaussianMixture
```

```python
gmm = GaussianMixture(n_components=3)
```

```python
labels = gmm.fit_predict(x)
```

```python
labels
```

```python
x['labels'] = labels
```

```python
sns.pairplot(x, hue='labels', palette='tab10')
```

### Build the PCA

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
x_new = x.drop('labels', axis=1)
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

pca = PCA()
x_pca = pca.fit_transform(x_scaled)
```

```python
x_pca = pd.DataFrame(x_pca)
x_pca['labels'] = labels
```

```python
sns.pairplot(x_pca, hue='labels', palette='tab10')
```
