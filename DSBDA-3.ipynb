### Descriptive Statistics - Measures of Central Tendency and variability

### Provide summary statistics (mean, median, minimum, maximum, standard deviation) for a dataset (age, income etc.) with numeric variables grouped by one of the qualitative (categorical)variable

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
```

```python
df.describe()
```

```python
df['Gender'].value_counts()
```

```python
gr = df.groupby('Gender')
```

```python
gr.get_group('Male')
```

```python
gr_age = df.groupby('Age')
```

```python
gr_age.get_group(27)
```

```python
gr.agg(np.mean)
```

```python
gr.agg(np.max)
```

```python
gr.agg(np.min)
```

```python
gr.agg(np.std)
```

```python
gr.agg(np.median)
```

### Write a Python program to display some basic statistical details like percentile, mean, standard deviation etc. of the species of ‘Iris-setosa’, ‘Iris-versicolor’ and ‘Iris-verginica’ of iris.csv dataset.

```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

```python
gr = iris.groupby('species')
```

```python
gr.groups
```

```python
### Setosa
gr.get_group('setosa').describe()
```

```python
# Versicolor
gr.get_group('versicolor').describe()
```

```python
# Verginica
gr.get_group('virginica').describe()
```
